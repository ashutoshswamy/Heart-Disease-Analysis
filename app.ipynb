### Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the dataset

```python
data = pd.read_csv("./heart.csv")
```

### Display the first 5 rows of the dataset

```python
data.head(5)
```

### Display the last 5 rows of the dataset

```python
data.tail(5)
```

### Display the shape of the dataset

```python
data.shape
```

```python
print("No. of rows:", data.shape[0])
print("No. of columns:", data.shape[1])
```

### Get info about the dataset

```python
data.info()
```

### Check for null values in the dataset

```python
data.isnull().sum()
```

### Check for duplicates in the dataset and drop them

```python
duplicate_data = data.duplicated().any()
print(duplicate_data)
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

### Get overall stats about the dataset

```python
data.describe()
```

### Correlation Matrix

```python
plt.figure(figsize = (15, 7))
sns.heatmap(data.corr(), annot = True)
plt.title("Correlation Matrix")
```

### Check how many people have heart disease and how many people don't have heart disease

```python
data.columns
```

```python
data["target"].value_counts()
```

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = data["target"])
plt.title("Countplot for people with heart disease and for people without heart disease")
```

### Find the number of males and females in the dataset

```python
data.columns
```

```python
data["sex"].value_counts()
```

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = data["sex"])
plt.xticks([0, 1], ["Female", "Male"])
plt.title("Number of males and females")
```

### Find Gender Distribution with respect to the `target` variable

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = "sex", hue = "target", data = data)
plt.xticks([0, 1], ["Female", "Male"])
plt.title("Gender Distribution")
plt.legend(labels = ["Non-Diseased", "Diseased"])
```

### Display the age distribution in the dataset

```python
plt.figure(figsize = (15, 7))
sns.histplot(data["age"], bins = 20)
plt.title("Age Distribution")
```

### Check which type of chest pain is common in the dataset

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = data["cp"])
plt.xticks([0, 1, 2, 3], ["Typical Anigna", "A-Typical Angina", "Non-Anginal Pain", "Asymptomatic"])
plt.title("Distribution of type of chest pain")
```

### Chest pain type distribution with respect to the `target` variable

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = "cp", hue = "target", data = data)
plt.title("Chest pain type Distribution")
plt.legend(labels = ["Non-Diseased", "Diseased"])
```

### Fasting Blood Sugar distribution with respect to the `target` variable

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = "fbs", hue = "target", data  = data)
plt.title("FBS Distribution")
plt.legend(labels = ["Non-Diseased", "Diseased"])
```

### Check Resting Blood Pressure Distribution

```python
plt.figure(figsize = (15, 7))
data["trestbps"].hist()
```

### Compare Resting Blood Pressure as per the sex

```python
plt.figure(figsize = (15, 7))
g = sns.FacetGrid(data, hue = "sex", aspect = 4)
g.map(sns.kdeplot, "trestbps", fill = True)
plt.title("Resting Blood Pressure as per the sex")
plt.legend(labels = ["Male", "Female"])
```

### Check Serum Cholesterol distribution

```python
plt.figure(figsize = (15, 7))
data["chol"].hist()
plt.title("Serum Cholesterol Distribution")
```

### Plot Continuous Variables

```python
data.columns
```

```python
categories_val = []
count_val = []

for column in data.columns:
    if data[column].nunique() <= 10:
        categories_val.append(column)
    else:
        count_val.append(column)
```

```python
categories_val
```

```python
count_val
```

```python
data.hist(count_val, figsize = (15, 7))
plt.tight_layout()
```
